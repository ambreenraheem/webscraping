# World Population Data scraped from Worldometers.com

### We need to use the requests library to download the HTML content of the page. The requests library allows us to send HTTP requests and receive responses from web servers. It is a powerful and easy-to-use library that simplifies the process of making HTTP requests in Python.

```python
# step 1: Import the required libraries
import pandas as pd

# step 2: Load the URL
url2 = 'https://www.worldometers.info/world-population/population-by-country/'
```

```python
# step 3: Fetch the HTML content with headers and read the HTML table into a DataFrame
import requests
# headers to avoid being blocked
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
```

```python
# headers are important to avoid being blocked by the server
response = requests.get(url2, headers=headers)
response.raise_for_status()  # Raise an error for bad status codes
```

```python
df2 = pd.read_html(response.text)
```

```python
# step 4: Check the number of tables
print(f"Number of tables found: {len(df2)}")
# step 5: Display the first few rows of the first table
df2[0].head()
```
